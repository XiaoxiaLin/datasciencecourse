# Marginal and conditional distribution

Let us illustrate the concepts of marginal and conditional distributions. 

The **marginal distribution** corresponds to the collapse/projection of the joint probability over one or several variables, e.g.

$$p(x_1) = \int p(x_1,x_2) dx_2$$


Let us check this concept in practice

```python
#GAUSSIAN
import numpy as np
import copy
from sklearn.base import BaseEstimator

class Gaussian(BaseEstimator):
    def __init__(self,mean=None,cov=None):
        self.mean = mean 
        self.cov =  cov
        if cov!=None:
            self.icov = np.linalg.pinv(self.cov)
    
    def fit_(self,X,y=None):
        self.mean = np.mean(X,axis=0)
        self.cov = np.cov(X,rowvar=False,bias = True)
        self.icov = np.linalg.pinv(self.cov)
    
    def fit(self,X,y):
        '''This is needed for compatibility with grid search CV and pipeline.
        When using this for modeling with data of just one class use fit_() method. 
        In that case target parameter is not needed.
        '''
        Xt=X[y==self.target,:]
        self.fit_(Xt)
    
    def predict(self,X):
        err = X-self.mean
        return np.where(-0.5*np.sum(np.dot(err,self.icov)*err,axis=1)>self.thr,1.,-1.)
    
    def predict_proba(self,X):
        err = X-self.mean
        return  np.exp(-0.5*np.sum(np.dot(err,self.icov)*err,axis=1))/np.sqrt(np.power((2*np.pi),self.mean.shape[0])*np.linalg.det(self.cov))
    
print 'DONE'
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
a = 0.9
K=np.array([[1,a],[a,1]])
g = Gaussian(cov=K, mean=np.array([0,0]))

xx = np.linspace(-3.,3.,100)
yy = np.linspace(-3.,3.,100)
[XX,YY] = np.meshgrid(xx,yy)
print np.c_[XX.ravel(),YY.ravel()].shape
Z = g.predict_proba(np.c_[XX.ravel(),YY.ravel()])

Z = Z.reshape(XX.shape)
plt.imshow(Z, interpolation='bilinear', origin='lower', extent=(XX.min(), XX.max(), YY.min(), YY.max()),alpha=0.3, vmin=0, vmax=1)
plt.contour(XX,YY,Z)
fig = plt.gcf()
fig.set_size_inches(9,9)
ax = plt.gca()
ax.set_xlabel('$x_1$', size = 26)
ax.set_ylabel('$x_2$', size = 26)
```

# Gaussian Processes

A Gaussian process is the generalization of the Gaussian distribution to "infinitely many variables".

Informally: An infinite long vector $\approx$ a function.

Definition: A Gaussian Process is a collection of random variables, any finite number of which have (consistent) Gaussian distribution.

A Gaussian distribution if fully specified by the mean $\mu$ and the covariance matrix $\Sigma$,

$$f = (f_1,\dots,f_N)^T \sim \mathcal{N}(\mu,\Sigma)$$
with indices $i = 1\dots N$

A Gaussian process is fully specified by a mean function $m(x)$ and a covariance function $k(x,x')$,

$$f(x) \sim \mathcal{GP} (m(x),k(x,x'))$$
with indices $x$.

Gaussian process especifies a distribution over functions.

Thinking of a Gaussian Process as a Gaussian distribution with an infinitely long mean vector and covariance matrix may seem impractical...

luckily we are saved by the marginalization property, that states that

$$\int p(x,y) d(y) = p(x)$$

And for Gaussians

$$p(x,y) = \mathcal{N}(\left[\begin{matrix}a\\ b \end{matrix}\right],\left[\begin{matrix} A & B\\ B^T & C\\ \end{matrix}\right]) \implies p(x) = \mathcal{N}(a,A)$$

If you ask finite dimensional questions to the gaussian process you get a finite dimensional answer. And this finite dimensional answer depends only on the finite dimensional set and does not depend on the infinitely long rest. 

Let us now draw random samples from a Gaussian Process. Let us start defining the Gaussian process as follows

$$m(x) = 0$$

$$k(x,x') = \sigma_f e^{-1/2 (x-x')^TD^{-1}(x-x')} + \sigma_n \delta(x)$$

In this particular case the covariance also resembles a Gaussian but it can be anything.

Now, how do I sample a function from my gaussian process. If we consider a function evaluated at certain points $x_1\dots x_N$, i.e. ${\bf f} = (f(x_1),\dots,f(x_N))$
$${\bf f}\sim \mathcal{N}(0,\Sigma)$$
where $\Sigma_{ij} = k(x_i,x_j)$.



Let me illustrate how the sampling works.

+ Get some $x$ at random
+ Compute the covariance matrix
+ Sample the normal distribution

```python
l=0.01
x = np.random.rand(100,1)         # 100 random x
k = np.exp(-0.5*(x-x.T)**2/l)

f = np.random.multivariate_normal(np.zeros(x.shape).ravel(),k,1)

print f
plt.plot(x.ravel(),f[0],'ro')
```

# Sequential generation

Factorize the joint distribution

Example: $p(x,y,z) = p(z)p(y|z)p(x|y,z) =  p(x,y,z) = p(z)p(x|z)p(y|x,z), $  it doesn't depends on the sequence

$$p(f_1,\dots,f_n|x_1,\dots,x_n) = \prod_{i=1}^n p(f_i|f_{i-1},\dots,f_1,x_i,x_{i-1},\dots,x_1)$$

and with a Gaussian 

$$p(x,y) = \mathcal{N}(\left[\begin{matrix}a\\ b \end{matrix}\right],\left[\begin{matrix} A & B\\ B^T & C\\ \end{matrix}\right]) \implies p(x|y) = \mathcal{N}(a+BC^{-1}(y-b),A-BC^{-1}B^T)$$

Observe that the conditional covariance matrix is given by the **Schur complement** of the non conditioned block.


$$p(x,y) = \mathcal{N}(\left[\begin{matrix}a\\ b \end{matrix}\right],\left[\begin{matrix} A & B\\ B^T & C\\ \end{matrix}\right]) \implies p(y|x) = \mathcal{N}(b + B^T A^{-1} (x-a),C - B^T A^{-1}B)$$

```python
__verbose__=False
ele=0.001
x = np.random.rand(100,1)
x = np.linspace(0,1,100)[:,np.newaxis]

k = np.exp(-0.5*(x-x.T)**2/ele)     # covariance function that imposes smoothness of our samples
print "k.shape:",k.shape

f = np.zeros(x.shape)
f[0] = np.random.normal()

print f[0]
for i in xrange(1,x.shape[0]):
    A = k[:i,:i]
    B = k[:i,i][:,np.newaxis]
    #C is 1 in this application since it is k(i,i)
    
    Ai = np.linalg.inv(A+1e-4*np.eye(A.shape[0]))
    mu = np.dot(np.dot(B.T,Ai),f[:i])
    cov = 1.-np.dot(np.dot(B.T,Ai),B)
    f[i] = np.random.normal(mu,cov)   # generate random sample with the mu and cov
    
    if __verbose__:
        print 'mu: '+ str(mu)
        print 'cov: '+str(cov)
    
plt.plot(x.ravel(),f,'ro')
```

# Prediction

This is the same process we may use for prediction purposes.

Recall that

$$p(x,y) = \mathcal{N}(\left[\begin{matrix}a\\ b \end{matrix}\right],\left[\begin{matrix} A & B\\ B^T & C\\ \end{matrix}\right]) \implies p(x|y) = \mathcal{N}(a+BC^{-1}(y-b),A-BC^{-1}B^T)$$

Now consider that we look for the probability of test data given training data. Remember that our **prior** is that ${\bf f}\sim \mathcal{GP}(0,k(x,x'))$.

If we assume that our data can be noisy $y = f(x) + \epsilon$, where $\epsilon \sim \mathcal{N}(0,\sigma^2_{noise})$, then our prior is modified to accomodate this new term, 
${\bf f}\sim \mathcal{GP}(0,k(x_p,x_q)+\sigma^2_{noise}\delta_{pq})$, where $\delta_{pq} = \left\{ \begin{align} 1& \quad \text{if} \; p=q\\ 0 & \quad \text{otherwise}\end{align}\right.$

$$\left[\begin{matrix}y\\f^*\end{matrix}\right] \sim \mathcal{N}(\left[\begin{matrix}0\\0\end{matrix}\right],
\left[\begin{matrix}K({\bf x},{\bf x})+\sigma_{noise}I & K({\bf x},x^*)\\K({\bf x},x^*)^T& K(x^*,x^*)\end{matrix}\right])$$



Let us generate our well known Chevyshev's basis polynomial.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def polyval(coefs, x):
    res = coefs[0]*np.ones((1,x.shape[0]))
    for i in xrange(1,len(coefs)):
        res = res + coefs[i]*np.power(x,i)
    return res
        

def chebys (coefs, x):
    pol = {}
    pol[0] = [1][::-1]
    pol[1] = [1, 0][::-1]
    pol[2] = [2, 0, -1][::-1]
    pol[3] = [4, 0, -3, 0][::-1]
    pol[4] = [8, 0, -8, 0, 1][::-1]
    pol[5] = [16, 0, -20, 0, 5, 0][::-1]
    pol[6] = [32, 0, -48, 0, 18, 0, -1][::-1]
    pol[7] = [64, 0, -112, 0, 56, 0, -7, 0][::-1]
    pol[8] = [128, 0, -256, 0, 160, 0, -32, 0, 1][::-1]
    pol[9] = [256, 0, -576, 0, 432, 0, -120, 0, 9, 0][::-1]
    pol[10] = [512, 0, -1280, 0, 1120, 0, -400, 0, 50, 0, -1][::-1]
    pol[11] = [1024, 0, -2816, 0, 2816, 0, -1232, 0, 220, 0, -11, 0][::-1]

    res = np.zeros((1,x.shape[0]))

    for i in xrange(len(coefs)):
        res= res + coefs[i]*polyval(pol[i],x)
    return res

        
x = np.linspace(0,1,100)

%matplotlib inline
import matplotlib.pyplot as plt

for i in xrange(11):
    co = np.zeros((11,))
    co[i]=1.
    plt.plot(x,chebys(co,x).ravel())
    
#Create a polynomial with noise
np.random.seed(42)
coefs = np.random.random(11)

x = np.linspace(0,1,100)

vals=chebys(coefs,x).ravel()
#plt.plot(x,vals)

N=10
idx = np.random.randint(0,100,N)
data = vals[idx]+0.5*np.random.normal(size=len(idx))
plt.figure()
plt.plot(x[idx],data,'ro')

x_train = x[idx][:,np.newaxis]
y_train = data[:,np.newaxis]
```

And now adjust the Gaussian process.

Modeling the problem accounts for stablishing the likelihood and the prior functions,

Gaussian likelihood

$$y|x,f(x),M_t \sim \mathcal{N}({\bf f},\sigma_{noise}I)$$

Zero mean Gaussian Process prior

$$f(x)|M_t \sim \mathcal{N}(0,k(x,x'))$$

which leads to a Gaussian process posterior 

$$f(x)|{\bf x},{\bf y},M_t \sim  \mathcal{GP} (m_{post}(x), k_{post}(x,x'))$$

where

$$m_{post}(x) = k(x,{\bf x})[K({\bf x},{\bf x})+\sigma_{noise}I]^{-1}{\bf y}$$

$$k_{post}(x,x') = k(x,x')-k(x,{\bf x})[K({\bf x},{\bf x})+\sigma^2_{noise}I]^{-1}k({\bf x},x')$$

and a Gaussian predictive distribution 

$$y^*|x^*,y,x,M_t \sim \mathcal{N}(k(x^*,{\bf x})[K({\bf x},{\bf x})+\sigma^2_{noise}I]^{-1}{\bf y} , k(x^*,x^*)+-k(x^*,{\bf x})[K({\bf x},{\bf x})+\sigma_{noise}I]^{-1}k({\bf x},x^*)$$


```python
__verbose__=False
l=0.01               # landscale: structure of the covariance matrix. if l big, very correlated.
sigma_noise=1e-4     # certancy of my data (points)
A = np.exp(-0.5*(x_train-x_train.T)**2/l)
Ai = np.linalg.inv(A+sigma_noise*np.eye(A.shape[0]))

x_test = np.linspace(0,1,100)

f = np.zeros(x_test.shape)
f_min = np.zeros(x_test.shape)
f_max = np.zeros(x_test.shape)
for i in xrange(0,x_test.shape[0]):
    B = np.exp(-0.5*(x_train-x_test[i].T)**2/l)
    mu = np.dot(np.dot(B.T,Ai),y_train)
    cov = 1.-np.dot(np.dot(B.T,Ai),B)
    f[i] = mu
    f_max[i] = mu + 2*np.sqrt(cov)
    f_min[i] = mu - 2*np.sqrt(cov)

    if __verbose__:
        print 'mu: '+ str(mu)
        print 'cov: '+str(cov)

plt.plot(x_train.ravel(),y_train,'ro')
plt.plot(x_test.ravel(),f,'r-')
plt.fill_between(x_test, f_min, f_max, facecolor='lightgray', interpolate=True)
```

Let us encapsulate this code in a class for the sake of completeness.

```python
#GAUSSIAN PROCESS

import numpy as np
import copy
from sklearn.base import BaseEstimator
import scipy.spatial.distance as distance

class GaussianProcess(BaseEstimator):
    def __init__(self,l=None,sigma_noise=0.01):
        '''l: is a vector of the scale for each dimension'''
        self.sigma_noise=sigma_noise
        self.l=l
    
    def fit(self,X,y):
        '''This is needed for compatibility with grid search CV and pipeline.
        When using this for modeling with data of just one class use fit_() method. 
        In that case target parameter is not needed.
        '''
        self.X=X
        self.y=y
        d = distance.cdist(self.X,self.X,metric='seuclidean',V=self.l)
        A=np.exp(-0.5*d*d)
        self.Ai = np.linalg.inv(A+self.sigma_noise*np.eye(A.shape[0]))
    
    def predict(self,X):
        d = distance.cdist(X,self.X,metric='seuclidean',V=self.l)
        B=np.exp(-0.5*d*d).T
        d = distance.cdist(X,X,metric='seuclidean',V=self.l)
        C=np.exp(-0.5*d*d)
        self.mu = np.dot(np.dot(B.T,self.Ai),self.y)
        return self.mu
    
    def predict_proba(self,X):
        d = distance.cdist(X,self.X,metric='seuclidean',V=self.l)
        B=np.exp(-0.5*d*d).T
        d = distance.cdist(X,X,metric='seuclidean',V=self.l)
        C=np.exp(-0.5*d*d)
        self.mu = np.dot(np.dot(B.T,self.Ai),self.y)
        self.cov = C - np.dot(np.dot(B.T,self.Ai),B)
        return  self.mu,np.diag(self.cov)
    
print 'DONE'
```

# Kernels in Gaussian Processes

The covariance matrix aka kernel models the relationship or similarity among samples in our dataset. We have already seen one of these kernel functions the **Gaussian kernel, Squared exponential or Radial Basis Function** defined as

$$k(x,x') = \sigma_s \exp (-\frac{1}{2}\frac{\|x-x'\|^2}{l})$$

where $\sigma_s$ accounts for the magnitude of the variation is the and $l$ is the characteristic length scale.

Let us refactor the code to allow for our own definition of covariances,

```python
#GAUSSIAN PROCESS
import numpy as np
import copy
from sklearn.base import BaseEstimator
import scipy.spatial.distance as distance


def gaussian_covariance(a,b,l,s):
    d = distance.cdist(a,b,metric='seuclidean',V=l)
    return s*np.exp(-0.5*d*d)

class GaussianProcess(BaseEstimator):
    def __init__(self,sigma_noise=0.01,covariance = None, arguments = None):
        '''l: is a vector of the scale for each dimension'''
        self.sigma_noise=sigma_noise
        self.covariance = covariance
        self.args = arguments
    
    def fit(self,X,y):
        self.X=X
        self.y=y
        A = self.covariance(self.X,self.X,**self.args)
        self.Ai = np.linalg.inv(A+self.sigma_noise*np.eye(A.shape[0]))
    
    def predict(self,X):
        B = self.covariance(X,self.X,**self.args).T
        C = self.covariance(X,X,**self.args)
        self.mu = np.dot(np.dot(B.T,self.Ai),self.y)
        return self.mu
    
    def predict_proba(self,X):     
        B = self.covariance(X,self.X,**self.args).T
        C = self.covariance(X,X,**self.args)
        self.mu = np.dot(np.dot(B.T,self.Ai),self.y)
        self.cov = C - np.dot(np.dot(B.T,self.Ai),B)
        return  self.mu,np.diag(self.cov)
    
print 'DONE'
```

```python
gp = GaussianProcess(sigma_noise=1e-4,covariance=gaussian_covariance, arguments = {'l':np.array([0.01]),'s':1.})
gp.fit(x_train,y_train)
f,c = gp.predict_proba(x_test[:,np.newaxis])
f_max = f.ravel() + 2*np.sqrt(c)
f_min = f.ravel() - 2*np.sqrt(c)

plt.plot(x_train.ravel(),y_train,'ro')
plt.plot(x_test.ravel(),f,'k-')
plt.fill_between(x_test, f_min, f_max, facecolor='lightgray', interpolate=True)
```

## Other kind of kernels

Kernels are going to model the kind of trends or patterns we are expecting in the data. There are many different kernels:

+ **Linear covariance: ** models linear relationships in the data.

$$k(x,x') = \sigma_1 {\bf x}{\bf x'^T} + \sigma_2$$


+ **Exponential sine covariance: ** models periodic behaviors.

$$k(x,x') = \exp(- \frac{2 \sin^2(\pi/p(x-x'))}{l})$$


```python
def linear_covariance(a,b,sigma1,sigma2):
    return sigma1*np.dot(a,b.T)+sigma2

def exponential_sine_covariance(a,b,l,p):    
    '''l: length scale p: periodicity'''
    r = a-b.T
    return np.exp(-2*np.sin(np.pi/p*r)**2/l)
```

```python
gp = GaussianProcess(sigma_noise=1e-1,covariance=linear_covariance, arguments = {'sigma1':10.,'sigma2':1000.})
gp.fit(x_train,y_train)
f,c = gp.predict_proba(x_test[:,np.newaxis])
f_max = f.ravel() + 2*np.sqrt(c)
f_min = f.ravel() - 2*np.sqrt(c)

plt.plot(x_train.ravel(),y_train,'ro')
plt.plot(x_test.ravel(),f,'k-')
plt.fill_between(x_test, f_min, f_max, facecolor='lightgray', interpolate=True)
```

```python
gp = GaussianProcess(sigma_noise=1e-2,covariance=exponential_sine_covariance, arguments = {'l':np.array([1.]),'p':0.006})
gp.fit(x_train,y_train)
f,c = gp.predict_proba(x_test[:,np.newaxis])
f_max = f.ravel() + 2*np.sqrt(c)
f_min = f.ravel() - 2*np.sqrt(c)

plt.plot(x_train.ravel(),y_train,'ro')
plt.plot(x_test.ravel(),f,'r-')
plt.fill_between(x_test, f_min, f_max, facecolor='lightgray', interpolate=True)
```

## Composition of kernels

Additionally, if $k_1$ and $k_2$ are two valid kernels, then, 

$$k = k_1 + k_2$$

and 

$$k = k_1 \cdot k_2$$ 

are also valid kernels.

```python
#GAUSSIAN PROCESS
import numpy as np
import copy
from sklearn.base import BaseEstimator
import scipy.spatial.distance as distance

class GaussianProcess(BaseEstimator):
    def __init__(self,sigma_noise=0.01,covariance = None, arguments = None):
        '''l: is a vector of the scale for each dimension'''
        self.sigma_noise=sigma_noise
        self.covariance = covariance
        self.args = arguments
    
    def fit(self,X,y):
        self.X=X
        self.y=y
        A = self.covariance(self.X,self.X,**self.args)
        self.Ai = np.linalg.inv(A+self.sigma_noise*np.eye(A.shape[0]))
  
    
    def predict(self,X):
        B = self.covariance(X,self.X,**self.args).T
        C = self.covariance(X,X,**self.args)
        self.mu = np.dot(np.dot(B.T,self.Ai),self.y)  
        return self.mu
    
    def predict_proba(self,X):     
        B = self.covariance(X,self.X,**self.args).T
        C = self.covariance(X,X,**self.args)
        self.mu = np.dot(np.dot(B.T,self.Ai),self.y)
        self.cov = C - np.dot(np.dot(B.T,self.Ai),B)
        return  self.mu,np.diag(self.cov)
    
print 'DONE'
```

```python
def linear_covariance(a,b,sigma1,sigma2):
    return sigma1*np.dot(a,b.T)+sigma2

#  Exp-Sine-Squared kernel
def exponential_sine_covariance(a,b,l,p):    
    '''l: length scale p: periodicity'''
    r = a-b.T
    return np.exp(-2*np.sin(np.pi/p*r)**2/l)**2

def gaussian_covariance(a,b,l,s):                  
    d = distance.cdist(a,b,metric='seuclidean')    
    return s*np.exp(-0.5*d*d)   


def mykernel(a,b,l_sine,p_sine,l_g_sine,s_g_sine,l_g,s_g):
    cov1= gaussian_covariance(a,b,l = l_g,s = s_g)
    cov2= gaussian_covariance(a,b,l = l_g_sine, s = s_g_sine) + exponential_sine_covariance(a,b,l=l_sine,p=p_sine)
    cov3 = linear_covariance(a,b,1.,500.)
    return cov1+cov2 +cov3 
```
