### 2 dimension plot

```python
# Plot x^2+y^2

# This is for Linux KDE
#import matplotlib
#matplotlib.use('Qt5Agg')

# This is general
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

# Surface to plot
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Surface plot 3D + contour plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                linewidth=0, antialiased=False)
ax.contour(X,Y,Z,30,zdir='z',offset=0)

plt.show()
```

## The exercice

1. plot the following function:

    \begin{equation}
      f(x_1,x_2) = x^2_1 (4- 2.1x^2_1+\frac{1}{3}x^4_1)+x_1x_2+x^2_2(-4+4x^2_2)
    \end{equation}

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Surface to plot
X1 = np.arange(-2, 2, 0.05)
X2 = np.arange(-1, 1, 0.05)
X1, X2 = np.meshgrid(X1, X2)
```

```python
z_l1 = X1**2
z_l2 = 4-2.1*X1**2+(1./3.)*X1**4
z_m =X1*X2
```

```python
z_r1 = X2**2
z_r2 = -4+4*X2**2
```

```python
#vectorized equation

Z = np.multiply(z_l1, z_l2)+z_m+np.multiply(z_r1,z_r2)
```

```python
%pylab inline

# Surface plot 3D + contour plot
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,  cmap=cm.jet,
                linewidth=0, antialiased=True)
ax.contour(X1,X2,Z,10,zdir='z',offset=-1)
ax.view_init(azim=60)


plt.show()
```
