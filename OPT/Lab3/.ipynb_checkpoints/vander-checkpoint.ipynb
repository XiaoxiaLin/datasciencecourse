```python

import numpy as np

for n in range (2,51):
    alph=np.zeros(n);
    for j in range (1,n+1):
        alph[j-1] = 1-2.*(j-1)/(n-1)
    
    V=np.zeros((n,n))
    for i in range (0,n):
        for j in range (0,n):
            V[i,j] = alph[j]**i
    print V 
    kV=np.linalg.norm(V,2)*np.linalg.norm(np.linalg.inv(V),2)
    print n,kV
    

```
