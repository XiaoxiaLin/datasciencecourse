```python
# One dimensional case 
# Plot X**3-2*X+2

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# _Function to plot
X = np.arange(-2, 2, 0.025)
Y = X**3-2*X+2; 

def F(X):
    return X**3-2*X+2

def DF (X):
    return  3*(X**2) -2

def DDF (X):
    return 6*X

# calculate the points x* that satisfy f'(x)= 0 

import sympy
s = []
x = sympy.symbols('x')
s = sympy.solve(3*(x**2)-2,x)
s1, s2 = s[0], s[1]
print 'Solution to dy(x)=0 are: [s1,s2]=', s,', where we have maximum and minimum'

# compute the second derivate in these two points to see whether is maximum or minimum

x1, x2 = DDF(s1), DDF(s2)
if DDF(s1)<0 and DDF(s2)>0:
    print '\nddf(s1)=',x1,' The secondary derivative at s1 is <0, than s1=', s1, 'is maximum'
    print 'ddf(s2)=',x2,' The secondary derivative at s2 is >0, than s2=', s[1], 'is mimimum'

# let's perform a Taylor expansion arount the point x* 

print '\nf(s1)=', F(s1)
print 'f(s2)=', F(s2)

G = F(s2)+ DF(s2)*(X-s2) + (1./2)*DDF(s2)*(X-s2)**2

plt.plot(X,Y)
plt.plot(X,G)
```

```python
# Two dimensional case
# Plot x^2+y^2

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

# Surface to plot
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Surface plot 3D + contour plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                linewidth=0, antialiased=False)
ax.contour(X,Y,Z,30,zdir='z',offset=0)

plt.show()
```

```python
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


def f(x,y):
    return np.sin((x + y))
x = y = np.arange(-5, 5, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([f(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)


gx,gy = np.gradient(Z,0.05,0.05)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface( X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                linewidth=0, antialiased=False)
ax.contour(X,Y,Z,30,zdir='z',offset=0)
```
